**Exemplo de Regress√£o**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y = True)
plt.scatter(X[:, 12], y)
plt.show()
```

**Exemplo Alternativo de Regress√£o em rela√ß√£o ao *dataset load_boston*** 

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=500, n_features=1, noise=40, random_state=4)
plt.scatter(X, y)
plt.show()
```

**Exemplo de Classifica√ß√£o**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
X, y = load_breast_cancer(return_X_y = True)
plt.plot(X[:, 20][y==1], X[:, 27][y==1], "ro")
plt.plot(X[:, 20][y==0], X[:, 27][y==0], "gs")
plt.show()
```

**Exemplo de Aprendizado N√£o Supervisionado**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y = True)
plt.plot(X[:, 0], X[:, 1], "o")
plt.show()
```
